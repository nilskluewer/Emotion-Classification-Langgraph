```python
def create_knowledge_query_translation():

    prompt = ChatPromptTemplate(
        [
            ("human", "Helo"),
            ("ai", "Hallo wie gehts?"),
            ("human", "Danke {user}, mir gehts gut.")
        ]
    )

    prompt2 = ChatPromptTemplate(
        [
            ("human", "{context}"),
        ]
    )

    print("PROMPT2", prompt2)

    # TODO: Create llm m√ºssen wir mit langchain machen da wir mehre modelle nehmen wollen!
    llm = create_llm(
        model_name="gemini-1.5-pro-001",
        temperature=0.5,
        response_mime_type="application/json",
    )
    print("PROMPT", prompt)

    chain = prompt | llm

    return chain


if __name__ == "__main__":
    chain = create_knowledge_query_translation()
    result = chain.invoke({"user": "Nils"})
    print("\n ----- \n")
    print(result)
```
