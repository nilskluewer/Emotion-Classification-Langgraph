```python
def create_knowledge_query_translation():

    prompt = ChatPromptTemplate(
        [
            ("human", "Helo, ich bin {user}, wie heißt du?"),
            ("ai", "Hallo wie gehts? Ich heiße {ai_name}"),
            ("human", "Wie heißt du nochmal?")
        ]
    )

    template = prompt.invoke({
        "user": "Bob",
        "ai_name": "AI_infunction"
    })

    # TODO: Create llm müssen wir mit langchain machen da wir mehre modelle nehmen wollen!
    llm = create_llm(
        model_name="gemini-1.5-pro-001",
        temperature=0.5,
        response_mime_type="application/json",
    )
    print("PROMPT", prompt)

    chain = prompt | llm

    return chain


if __name__ == "__main__":
    chain = create_knowledge_query_translation()
    result = chain.invoke({"user": "Nils", "ai_name": "AI_outfunction"})
    print("\n ----- \n")
    print(result)
```
